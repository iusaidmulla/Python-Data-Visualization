```python
# library and dataset
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Create data
df=pd.DataFrame({'x': range(1,101), 'y': np.random.randn(100)*15+range(1,101)})
```

```python
# plot with matplotlib
plt.plot( 'x', 'y', data=df, color='red')
plt.show()
```

```python
# plot with matplotlib
plt.plot( df.x, df.y, color='red')
plt.show()
```

```python
# plot with matplotlib
plt.plot( df.x, df.y, marker = 'o',color='red')
plt.show()
```

```python
# plot with matplotlib
plt.scatter( df.x, df.y, color='red')
plt.show()
```

```python
plt.scatter(df.x, df.y, color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Scatter Plot')
plt.show()
```

```python
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars))
```

```python
plt.bar(y_pos, height)
plt.xticks(y_pos, bars)
plt.show()
```

```python
#plt.bar(y_pos, height)
plt.barh(y_pos, height)
plt.yticks(y_pos, bars)
plt.show()
```

```python
#different color for each bar
plt.bar(y_pos, height, color=['black', 'red', 'green', 'blue', 'cyan'])
plt.xticks(y_pos, bars)
plt.show()
```

```python
#different color for each bar
plt.bar(y_pos, height, color=['black', 'red', 'green', 'blue', 'cyan'], edgecolor = 'blue')
plt.xticks(y_pos, bars)
plt.show()
```
