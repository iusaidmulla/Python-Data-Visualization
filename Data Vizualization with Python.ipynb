## Data Vizualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
yield_apples = [0.897, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
plt.plot(yield_apples) #line charts
plt.show()
```

```python
yield_apples = [0.897, 0.91, 0.919, 0.926, 0.929, 0.931]
years = [2010,2011,2012,2013,2014,2015]

plt.plot(years, yield_apples,color='green') #line charts

plt.title('Apple Yield 2010-15',size=15)
plt.xlabel('Years')
plt.ylabel('Yield Percentage')
plt.show()
```

```python
yield_apples = [0.897, 0.91, 0.919, 0.926, 0.929, 0.931]
yield_bananas = [0.962, 0.94, 0.918, 0.904, 0.899, 0.888]
years = [2010,2011,2012,2013,2014,2015]

plt.plot(years, yield_apples,color='red') #line charts
plt.plot(years, yield_bananas,color='yellow') #line charts

plt.title('Apple Yield 2010-15',size=20)
plt.xlabel('Years')
plt.ylabel('Yield Percentage')
plt.legend(['Apples','Bananas'])
plt.show()
```

```python
# plt.plot(years, yield_apples,color='red') #line charts
# plt.plot(years, yield_bananas,color='yellow') #line charts


plt.figure(figsize=(8,8))

plt.plot(years, yield_apples,color='red', marker = '*', ls= ':', lw = 2) 
plt.plot(years, yield_bananas,color='blue', marker = 'v', ls= '--', lw = 2) 


plt.title('Apple Yield 2010-15',size=20)
plt.xlabel('Years')
plt.ylabel('Yield Percentage')
plt.legend(['Apples','Bananas'])
plt.show()
```

## Scatter Plots

```python
#by deafault some data available for us in these vizulaization libraries
#iris dataset
flowers_df = sns.load_dataset("iris")
flowers_df.head()
```

```python
flowers_df.species.unique()
```

```python
flowers_df['sepal_length']
```

```python
flowers_df['sepal_width']
```

```python
#using matplotlib
plt.figure(figsize = (12,8))
plt.scatter(flowers_df['sepal_length'], flowers_df['sepal_width'])
plt.show()
```

```python
#using seaborn
plt.figure(figsize = (8,6))

sns.scatterplot(x=flowers_df['sepal_length'], 
                y=flowers_df['sepal_width'],
                 hue=flowers_df['species'],s=100 )
plt.show()
```

## Histograms

```python
sns.set_style?
```

```python
sns.set_style("darkgrid")
plt.figure(figsize = (8,6))

plt.hist(flowers_df['sepal_width'])

plt.title("Distribution of Sepal Width",size=30)

plt.show()
```

```python
#bins
plt.figure(figsize = (8,6))

plt.hist(flowers_df['sepal_width'],bins=5)

plt.title("Distribution of Sepal Width",size=30)

plt.show()
```

```python
#Comparing two distributions of setosa,versicolor,virginica
setosa_width = flowers_df[flowers_df['species']=='setosa']['sepal_width']

versicolor_width = flowers_df[flowers_df['species']=='versicolor']['sepal_width']

plt.figure(figsize = (8,6))

plt.hist(setosa_width,bins=5, alpha = 0.4)
plt.hist(versicolor_width,bins=5,alpha = 0.2)

plt.show()
```

## Bar Charts

```python
years = range(2015,2021)
yield_apples = [0.897, 0.61, 0.93, 0.78, 0.929, 0.40]
yield_bananas = [0.962, 0.94, 0.918, 0.904, 0.899, 0.888]
```

```python
plt.bar(years,yield_apples,color ='brown')
plt.xticks(years,['Year 1','Year 2','Year 3','Year 4','Year 5','Year 6'])
plt.show()
```

```python
plt.bar(years,yield_apples)

plt.bar(years,yield_bananas, bottom = yield_apples)

plt.show()
```
